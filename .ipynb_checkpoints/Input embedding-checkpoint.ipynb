```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import matplotlib

df = pd.read_csv('train.csv')
df.head(10)
```

```python
X_text=df.Sentence
y_train=df.NER
len(X_text)
```

```python
word_list=list(set(' '.join(X_text).split()))

if ('Pad' in word_list)==False:
    word_list.append('Pad')
    
word_dict = {w: i for i, w in enumerate(word_list)}
print('Vord dictionary size: ',len(word_dict))
```

## TF-IDF

```python
from sklearn.feature_extraction.text import CountVectorizer  
from sklearn.feature_extraction.text import TfidfTransformer 


vectorizer=CountVectorizer()
transformer=TfidfTransformer(smooth_idf=True) 

tfidf=transformer.fit_transform(vectorizer.fit_transform(X_text))
word=vectorizer.get_feature_names()
weight=tfidf.toarray()
print(word)
weight
```

## word-vec

```python
X_train=[sentence.split() for sentence in X_text]

# Get the words quantity in each sentence
sen_len=[len(sentence) for sentence in X_train]
max_sen_len=max(sen_len)
# plt.hist(sen_len,bins=100)
print('Maximum word length of sentence: ',max_sen_len)
```

```python
text_vec=[]
for s in X_train:
    sen_vec=[word_dict[w] for w in s]
    for i in range(max_sen_len-len(s)):
            sen_vec.append(word_dict['Pad'])
    text_vec.append(sen_vec)
    
len(text_vec),len(text_vec[0])
```
